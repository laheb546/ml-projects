```python
categorical_features_with_na=[feature for feature in dataset.columns if dataset[feature].isnull().sum()>0 and dataset[feature].dtypes=='str']
```

```python
categorical_features_with_na
for feature in categorical_features_with_na:
    print('{}:{} % missing values'.format(feature,np.round(dataset[feature].isnull().mean(),4)) )
```

```python
def replace_na_value(dataset,categorical_features_with_na):
    dataset[categorical_features_with_na]=dataset[categorical_features_with_na].fillna('MISSING')
    return dataset
dataset=replace_na_value(dataset,categorical_features_with_na)
```

```python
numerical_features_with_na=[feature for feature in dataset.columns if dataset[feature].isnull().sum()>0 and dataset[feature].dtypes!='str']
```

```python
numerical_features_with_na
for feature in numerical_features_with_na:
    print('{}  {} % missing values'.format(feature,np.round(dataset[numerical_features_with_na].isnull().mean(),5)))
```

```python
for feature in numerical_features_with_na:
    median_value=dataset[feature].median()
    dataset[feature+'nan']=np.where(dataset[feature].isnull(),1,0)
    dataset[feature]=dataset[feature].fillna(median_value)
```
